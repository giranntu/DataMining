```python
def loadDataSet(datapath):
    with open (datapath,'r') as f:
        lines = f.readlines()
    allsortList = []
    allitem = []
    for index, line in enumerate(lines):
        line=line.strip()
        line = line.split()
        setList={}
        sortList = []    
        for num in range(len(line)):
            if(num == 0):
                pass
            elif((num%2) != 0):
                if setList.has_key(line[num]):
                    setList[line[num]].append(line[num+1])
                else:
                    setList[line[num]] = [line[num+1]]
            elif((num%2) == 0):
                allitem.append(line[num])
        for key in sorted(setList.iterkeys()):
            sortList.append(setList[key])
        allsortList.append(sortList)
    allitem = set(allitem)
    allitem = list(allitem)
    return allsortList,allitem
```

```python
# 轉成 vertical format table
def listToTable(allsortList):
    l=[]
    for a in range(len(allsortList)):
        for b in allsortList[a]:
            tmp=[a]
            tmp.append(b)  
            l.append(tmp)
    allbolleanList = []

    for transaction in range(len(l)):
        bolleanList = []
        for item in allitem:
            if item in l[transaction][1]:
                bolleanList.append(True)
            else:
                bolleanList.append(False)
        allbolleanList.append(bolleanList)   
    
    itemDF = pd.DataFrame(allbolleanList,columns=allitem)
    rowname = ['candidateID','transactionID']
    ctpd = pd.DataFrame(l,columns=rowname)
    itemTable = ctpd.join(itemDF)
    return itemTable
```

```python
def support(itemTable,allitem,sup):
    transactionTmp = None
    supportItem = []
    countersup = []
    for num in range(len(allitem)):
        counter = 0
        for a,b in zip(itemTable['candidateID'].tolist(),itemTable[allitem[num]].tolist()):
            if b == True and a != transactionTmp:
                counter+=1
                transactionTmp = a
        if counter > sup:
            supportItem.append(allitem[num])
            countersup.append(counter)
    return supportItem,countersup
```

```python
def intersectionList(candidateIDList,aL,bL):
    booleanList = []
    counter = 0
    for a1,b1 in zip(aL,bL):
        if a1 == b1 & a1 == True:
            booleanList.append(True)
        else:
            booleanList.append(False)
    transactionTmp = None
    for a,b in zip(candidateIDList,booleanList):
        if b == True and a != transactionTmp:
            counter+=1
            transactionTmp = a
    return booleanList,counter
```
